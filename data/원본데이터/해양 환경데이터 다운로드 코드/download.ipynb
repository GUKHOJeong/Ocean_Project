```python
Physics_var_names={'so':'염분','thetao':'해저수온','uo':'해수유속(E)','vo':'해수유속(N)','zos':'해수면높이','mlotst':'혼합층 깊이'}
Physics_variables = ['so','thetao','uo','vo','zos','mlotst']
months=[1,7]
```

```python
shapefile_path = 'data\east_sea.shp'
gdf = gpd.read_file(shapefile_path)
gdf.set_crs(epsg=4326, inplace=True)
gdf_transformed = gdf.to_crs('EPSG:5179')
n=0
for year in range(2011,2026):
    
    for month in months:
        if year == 2025:
            nc_path = 'Ocean_Physics_data/2025-01-01_2025-02-01_Ocean_Physics.nc'
            ds = xr.open_dataset(nc_path)
            print(f'2025-01-01 파일 분석시작하겠습니다.')
            if month == 7:
                break
        elif month == 1:
            nc_path = f'Ocean_Physics_data/{year}-0{month}-01_{year}-06-30_Ocean_Physics.nc'
            ds = xr.open_dataset(nc_path)
            print(f'{year}_{month}파일 분석시작하겠습니다.')

        else:
            nc_path = f'Ocean_Physics_data/{year}-0{month}-01_{year}-12-31_Ocean_Physics.nc'
            ds = xr.open_dataset(nc_path)
            print(f'{year}_{month}파일 분석시작하겠습니다.')
        i=0
        for variable in Physics_variables:
            var_name=Physics_var_names[variable]
            print(f'현재 {var_name} 변수 진행중입니다.')
            data = ds[variable]
            data= data.rio.write_crs("EPSG:4326", inplace=True)
            target_crs = data .rio.crs
            if gdf_transformed.crs  != target_crs:
                gdf_transformed = gdf_transformed.to_crs(target_crs)
            data_clipped = data.rio.clip(gdf_transformed.geometry, gdf_transformed.crs, drop=False, invert=False)

            # 공간 차원(lat, lon)에 대해 평균 계산 (시간 차원은 유지)
            regional_mean_data = data_clipped.mean(dim=['latitude', 'longitude'], skipna=True)

            # 결과를 pandas DataFrame으로 변환
            name = f'mean_{var_name}'
            df_regional = regional_mean_data.to_dataframe(name=name)
            df_regional.drop(columns=['spatial_ref'],inplace=True)
            df_regional.reset_index(inplace=True)
            print(df_regional)
            if 'depth' in df_regional.columns:
                depth_bins = [0, 50, 100, 200, 300, 400, 500]
                # 레이블: 각 구간을 대표하는 이름
                depth_labels = [f'{var_name}_50m', f'{var_name}_100m', f'{var_name}_200m', f'{var_name}_300m', f'{var_name}_400m', f'{var_name}_500m']
                df_filtered = df_regional[(df_regional['depth'] > 0) & (df_regional['depth'] <= 500)].copy()
                df_filtered['depth_category'] = pd.cut(df_filtered['depth'],
                                                bins=depth_bins,
                                                labels=depth_labels,
                                                right=True)
                df_aggregated = df_filtered.groupby(['time', 'depth_category'], observed=False)[name].mean().reset_index()
                df= df_aggregated.pivot_table(index='time',          # 행 인덱스
                                                columns='depth_category', # 열로 만들 컬럼
                                                values=name)
                print(1)
            else:
                df =df_regional
            print(df)
            
            if i ==0:
                df_part=df
                i+=1
            else:
                df_part = pd.merge(df_part,df,how='left',on='time')
        if n == 0 :
            final_df =df_part
            print(final_df)
            n+=1
        else:
            final_df =   pd.concat([final_df,df_part]) 
final_df.to_csv('Physics.csv',encoding='utf-8-sig',index=False)
```
