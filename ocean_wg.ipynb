```python
col_fish=['가자미류', '청어', '새우', '대구', '도루묵', '기타어류']
col_fish = rate_ocean_final[col_fish]
col_catch=['총어획량 ']
col_catch=rate_ocean_final[col_catch]
col_salt=['염분_50m','염분_100m', '염분_200m', '염분_300m', '염분_400m', '염분_500m']
col_salt=rate_ocean_final[col_salt]
col_tem=['해저수온_100m','해저수온_200m', '해저수온_300m', '해저수온_400m', '해저수온_500m']
col_tem=rate_ocean_final[col_tem]
col_speed=['해수유속(E)_50m','해수유속(E)_100m', '해수유속(N)_50m', '해수유속(N)_100m']
col_speed=rate_ocean_final[col_speed]
col_mea=['mean_해수면높이','mean_혼합층 깊이']
col_mea=rate_ocean_final[col_mea]
col_clo=['클로로필_50m', '클로로필_100m', '클로로필_200m', '클로로필_300m','클로로필_400m', '클로로필_500m']
col_clo=rate_ocean_final[col_clo]
col_jsy=['질산염_50m', '질산염_100m', '질산염_200m', '질산염_300m']
col_jsy=rate_ocean_final[col_jsy]
col_o2=['용존 산소_50m', '용존 산소_300m']
col_o2=rate_ocean_final[col_o2]
col_avg=['평균 풍속(m/s)', '평균기압(hPa)', '평균 상대습도(%)', '평균 기온(°C)', '평균 최대 파고(m)','평균 파주기(sec)']
col_avg=rate_ocean_final[col_avg]




```

```python
plt.figure(figsize=(20, 10))
plt.plot(rate_ocean_final['Date'],rate_ocean_final['해저수온_100m'],label='해저수온_100m')
plt.plot(rate_ocean_final['Date'],rate_ocean_final['해저수온_200m'],label='해저수온_200m')
plt.plot(rate_ocean_final['Date'],rate_ocean_final['해저수온_300m'],label='해저수온_300m')
plt.plot(rate_ocean_final['Date'],rate_ocean_final['해저수온_400m'],label='해저수온_400m')
plt.plot(rate_ocean_final['Date'],rate_ocean_final['해저수온_500m'],label='해저수온_500m')
plt.legend()
plt.show()
```

```python
rate_ocean_final_copy=rate_ocean_final.copy().reset_index()
```

```python
fig, ax_1 = plt.subplots(figsize=(12, 8))
ax_1.set_xlabel('Date')
ax_1.set_ylabel('총어획량 ')
ax_1.plot(rate_ocean_final_copy['Date'], rate_ocean_final_copy['총어획량 '],color='orange')
ax_1.tick_params(axis='y', labelcolor='orange')

# Axis 2
ax_2 = ax_1.twinx()
ax_2.set_ylabel('Target')
ax_2.plot(rate_ocean_final_copy['Date'], rate_ocean_final_copy['Target'])
ax_2.tick_params(axis='y', labelcolor='blue')

plt.tight_layout()
plt.show()
```
